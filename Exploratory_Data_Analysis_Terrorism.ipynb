# CLEARING DATA

```python
dataset.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'State',
                       'region_txt':'Region','city':'City','latitude':'Latitude','longitude':'Longitude','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed',
                       'nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                       'weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
```

```python
dataset=dataset[['Year','Month','Day','Country','State','Region','City','Latitude','Longitude','AttackType','Killed',
               'Wounded','Target','Summary','Group','Target_type','Weapon_type','Motive']]
```

```python
dataset.columns
```

```python
dataset.head()
```

# HANDLING MISSING DATA

```python
dataset.isnull().sum()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(dataset.isnull(), yticklabels=False, cbar=False, cmap="plasma")
```

```python
dataset['Wounded'] = dataset['Wounded'].fillna(0)
dataset['Killed'] = dataset['Killed'].fillna(0)
dataset['Casualities'] = dataset['Killed'] + dataset['Wounded']
```

```python
dataset['State'].fillna("unknown", inplace = True)
dataset['City'].fillna("unknown", inplace = True)
```

```python
dataset.dropna(axis=1, inplace=True)
```

```python
dataset.isnull().sum()
```

# OBSERVATION

```python
print("Country with the most attacks:",dataset['Country'].value_counts().idxmax())
print("City with the most attacks:",dataset['City'].value_counts().index[1])
print("Region with the most attacks:",dataset['Region'].value_counts().idxmax())
print("Year with the most attacks:",dataset['Year'].value_counts().idxmax())
print("Month with the most attacks:",dataset['Month'].value_counts().idxmax())
print("Group with the most attacks:",dataset['Group'].value_counts().index[1])
print("Most Attack Types:",dataset['AttackType'].value_counts().idxmax())
```

# EXPLORATORY DATA ANALYSIS

## DATA VISUALISATION

```python
plt.rcParams["figure.figsize"] = 20,10
sns.barplot(x = dataset["Month"].value_counts()[:20].index, y=dataset["Month"].value_counts()[:20].values)
plt.ylabel("Number of Attacks", size=18)
plt.xticks(rotation=90)
plt.xlabel("Month of Attack", size=20)
plt.title(" ATTACKS BY MONTH ", size=25, fontweight="bold")
plt.show()
```

### Highest number of attacks are in the month of May but there is not much of difference in number of attacks by month.

```python
year = dataset['Year'].unique()
years_count = dataset['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))
sns.barplot(x = year,y = years_count,palette = "tab10")
plt.xticks(rotation = 50)
plt.xlabel('Year',size=20)
plt.ylabel('Number of Attacks',size=20)
plt.title('NUMBER OF ATTACKS YEAR WISE',size=25,fontweight="bold")
plt.show()
```

### Highest number of attacks are in the year 2014.

```python
df = dataset[['Year','Killed']].groupby(['Year']).sum()
fig, ax4 = plt.subplots(figsize=(20,8))
df.plot(kind='bar',alpha=0.7,ax=ax4)
plt.xticks(rotation = 50)
plt.ylabel("Number of people killed",size=20)
plt.xlabel('Year',fontsize=20)
plt.title("PEOPLE DIED IN ATTACK",size=25,fontweight="bold")
top_side = ax4.spines["top"]
top_side.set_visible(False)
right_side = ax4.spines["right"]
right_side.set_visible(False)
```

### Number of people died in attack is highest in the year 2014 and least in the 1970.

```python
plt.rcParams["figure.figsize"] = 17,13
sns.barplot(x = dataset["Country"].value_counts()[:20].index, y=dataset["Country"].value_counts()[:20].values)
plt.ylabel("Number of Attacks", size=20)
plt.xticks(rotation=90)
plt.xlabel("Country", size=20)
plt.title("COUNTRY WISE ATTACKS", size=25, fontweight="bold")
```

### Iraq is the country with highest number of attacks. 

```python
pd.crosstab(dataset.Year, dataset.Region).plot(kind='area',stacked=False,figsize=(20,10))
plt.ylabel('Numbers of Attacks',size=20)
plt.xlabel("Year",size=20)
plt.title('TERRORIST ACTIVITIES BY REGION',size=25,fontweight="bold")
plt.show()
```

```python
plt.rcParams["figure.figsize"] = 20, 8
sns.barplot(x = dataset["Group"].value_counts()[:20].index, y=dataset["Group"].value_counts()[:20].values)
plt.ylabel("Total Attack", size=20)
plt.xticks(rotation=90)
plt.xlabel("Attack Group", size=20)
plt.title("ATTACKS BY GROUP", size=25, fontweight="bold")
```

```python
dataset['City'].value_counts().to_frame().sort_values('City',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(20,10),color=(0.3, 0.7, 0.3, 0.8))
plt.xticks(rotation = 50)
plt.xlabel("City",fontsize=20)
plt.ylabel("Number of attack",fontsize=20)
plt.title("TOP 10 MOST EFFECTED CITIES",size=25,fontweight="bold")
plt.show()
```

```python
dataset[['AttackType','Killed']].groupby(["AttackType"],axis=0).sum().plot(kind='bar',figsize=(20,8),color=(0.8, 0.2, 0.4, 0.6))
plt.xticks(rotation=50)
plt.ylabel('Number of people killed',fontsize=20)
plt.xlabel('Attack type',fontsize=20)
plt.title("NUMBER OF PEOPLE KILLED IN DIFFERENT ATTACK TYPES",fontsize=25,fontweight="bold")
plt.show()
```

#### Maximum number of people is killed by Armed Assault followed by Bombing/Explosion.
